### 1. Demonstrate how to join a list of strings into a single sentence.

Here I used the `join()` method to combine words from a list into one sentence.

```python
words = ["Python", "is", "easy", "to", "learn"]
sentence = " ".join(words)

print(sentence)
```

### 2. Write a code snippet to remove duplicates from a list using a set.

A set only stores unique values, so duplicate items get removed.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))

print(unique_numbers)
```

### 3. How do you check if a specific key exists in a dictionary?

We can use the `in` keyword to check whether a key is present in a dictionary.

```python
student = {
    "name": "Harsh",
    "age": 20,
    "course": "Python"
}

if "age" in student:
    print("The key exists in the dictionary.")
else:
    print("The key does not exist.")
```

### 4. Explain the difference between `is` and `==`.

`==` checks if two values are equal.  
`is` checks if two variables are pointing to the exact same object in memory.

```python
list_one = [10, 20, 30]
list_two = [10, 20, 30]
list_three = list_one

print(list_one == list_two)   # values are same
print(list_one is list_two)   # different objects
print(list_one is list_three) # same object
```

### 5. Create a simple calculator using if-elif-else logic.

This calculator takes two numbers and an operator, then performs the matching operation.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Answer:", num1 + num2)
elif operator == "-":
    print("Answer:", num1 - num2)
elif operator == "*":
    print("Answer:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Answer:", num1 / num2)
    else:
        print("Cannot divide by zero.")
else:
    print("Invalid operator.")
```

### 6. Write a program to find the largest number in a list without using `max()`.

I started by assuming the first number is the largest, then compared the rest one by one.

```python
numbers = [45, 12, 89, 33, 67]
largest = numbers[0]

for number in numbers:
    if number > largest:
        largest = number

print("Largest number is:", largest)
```

### 7. Demonstrate how to use the `.strip()` method on a messy input string.

The `.strip()` method removes extra spaces from the beginning and end of a string.

```python
messy_text = "   hello python   "
clean_text = messy_text.strip()

print("Before:", messy_text)
print("After:", clean_text)
```

### 8. How do you convert a float to an integer? What happens to the decimal?

A float can be converted to an integer using `int()`. The decimal part is removed, not rounded.

```python
price = 45.89
converted_price = int(price)

print(converted_price)
```

### 9. Write a script that prints the multiplication table of a given number.

This program prints the table from 1 to 10 for the number entered by the user.

```python
number = int(input("Enter a number: "))

for i in range(1, 11):
    print(number, "x", i, "=", number * i)
```
