### 1. Explain the difference between list and tuple with a memory-usage example.

A list can be changed after creating it, but a tuple cannot be changed. Also, a tuple usually takes a little less memory than a list.

```python
import sys

numbers_list = [1, 2, 3, 4, 5]
numbers_tuple = (1, 2, 3, 4, 5)

print("list:", numbers_list)
print("tuple:", numbers_tuple)

print("list memory:", sys.getsizeof(numbers_list))
print("tuple memory:", sys.getsizeof(numbers_tuple))
```

### 2. How do you swap two variables in a single line without a third variable?

```python
a = 10
b = 20

print("before:", a, b)

a, b = b, a

print("after:", a, b)
```

### 3. Create a script to find the data type of every element in a mixed list.

```python
items = [25, "hello", 4.5, True, [1, 2, 3], (8, 9)]

for i in items:
    print(i, "is", type(i))
```

### 4. Write a program to check if a string is a palindrome.

```python
text = input("Enter a string: ")

text = text.lower().replace(" ", "")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

### 5. Demonstrate the difference between `/` and `//` operators.

`/` gives the exact division answer, while `//` gives the floor value.

```python
x = 7
y = 2

print("7 / 2 =", x / y)
print("7 // 2 =", x // y)
```

### 7. Write a script to count the number of vowels in a user-provided sentence.

```python
sentence = input("Enter a sentence: ")
vowels = "aeiouAEIOU"
vowel_count = 0

for letter in sentence:
    if letter in vowels:
        vowel_count += 1

print("vowels =", vowel_count)
```

### 8. Explain the concept of f-strings for variable interpolation.

F-strings are used to put variables directly inside a string using curly brackets.

```python
name = "Harsh"
age = 20

print(f"My name is {name} and I am {age} years old")
```

### 9. Create a list of numbers from 1 to 20 and print only the even ones.

```python
nums = list(range(1, 21))

for n in nums:
    if n % 2 == 0:
        print(n)
```

### 10. What is the difference between `range(5)` and `list(range(5))`?

`range(5)` gives a range object. `list(range(5))` changes it into a proper list.

```python
r = range(5)
lst = list(range(5))

print("range(5):", r)
print("list(range(5)):", lst)
```
