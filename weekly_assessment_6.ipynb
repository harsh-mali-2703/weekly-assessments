### 1. Write a recursive function for the Fibonacci sequence.

This function calls itself to find the Fibonacci number.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


print(fibonacci(6))
```

### 2. Explain "Memoization" and how it helps recursive functions.

Memoization means saving the answer of a function call so we do not have to calculate it again.

It helps recursive functions because some recursive programs repeat the same work many times. In Fibonacci, the same smaller numbers are calculated again and again, so memoization makes the function faster.

```python
def fibonacci_memo(n, memo=None):
    if memo is None:
        memo = {}

    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]


print(fibonacci_memo(10))
```

### 3. Use reduce() from functools to find the product of a list.

`reduce()` keeps applying the same operation to the list until one final answer is left.

```python
from functools import reduce

numbers = [2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)

print(product)
```

### 4. Write a function to check if a number is an "Armstrong Number."

An Armstrong number is equal to the sum of its digits raised to the power of the number of digits.

```python
def is_armstrong(num):
    digits = str(num)
    power = len(digits)
    total = 0

    for digit in digits:
        total += int(digit) ** power

    return total == num


print(is_armstrong(153))
print(is_armstrong(123))
```

### 5. Create a closure that remembers a base value and adds it to an input.

A closure can remember a value from the outer function even after the outer function has finished running.

```python
def make_adder(base):
    def add_number(value):
        return base + value

    return add_number


add_10 = make_adder(10)
print(add_10(5))
```

### 6. Explain the "LEGB" rule in Python scope.

LEGB is the order Python uses when it searches for a variable name.

- **Local**: inside the current function
- **Enclosing**: inside an outer function
- **Global**: in the main program file
- **Built-in**: Python's built-in names like `print`, `len`, and `sum`

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(x)

    inner()


outer()
```

### 7. Write a function that converts a decimal number to binary without bin().

This uses repeated division by 2 and stores the remainders.

```python
def decimal_to_binary(num):
    if num == 0:
        return "0"

    binary = ""

    while num > 0:
        remainder = num % 2
        binary = str(remainder) + binary
        num = num // 2

    return binary


print(decimal_to_binary(10))
```

### 8. Create a "Generator" function using yield to produce infinite even numbers.

A generator gives values one by one. Since this one has a `while True` loop, it can keep producing even numbers forever.

```python
def even_numbers():
    number = 0

    while True:
        yield number
        number += 2


evens = even_numbers()

print(next(evens))
print(next(evens))
print(next(evens))
print(next(evens))
```

### 9. How do you pass a function as an argument to another function?

In Python, a function can be passed like a normal variable. The second function can then call it.

```python
def square(num):
    return num * num


def apply_function(function, value):
    return function(value)


answer = apply_function(square, 5)
print(answer)
```

### 10. Write a function to validate a password.

The password should have at least 8 characters, 1 digit, and 1 symbol.

```python
def validate_password(password):
    if len(password) < 8:
        return False

    has_digit = False
    has_symbol = False

    for char in password:
        if char.isdigit():
            has_digit = True
        elif not char.isalnum():
            has_symbol = True

    return has_digit and has_symbol


print(validate_password("hello123"))
print(validate_password("hello@123"))
```
