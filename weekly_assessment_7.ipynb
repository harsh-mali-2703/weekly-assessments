### 1. Define a class `Car` with attributes and a method to display details.

```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_details(self):
        print(f"Car: {self.year} {self.brand} {self.model}")


car1 = Car("Toyota", "Innova", 2022)
car1.display_details()
```

### 2. Explain the purpose of the `__init__` method.

```python
# __init__ is a constructor method in Python.
# It runs automatically when we create an object.
# Mainly, it is used to give starting values to object attributes.

class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no


student1 = Student("Harsh", 7)
print(student1.name, student1.roll_no)
```

### 3. Create a subclass `ElectricCar` that inherits from `Car`.

```python
class ElectricCar(Car):
    def __init__(self, brand, model, year, battery_capacity):
        super().__init__(brand, model, year)
        self.battery_capacity = battery_capacity


tesla = ElectricCar("Tesla", "Model 3", 2024, "75 kWh")
tesla.display_details()
print("Battery:", tesla.battery_capacity)
```

### 4. Demonstrate "Method Overriding" in inheritance.

```python
class ElectricCar(Car):
    def __init__(self, brand, model, year, battery_capacity):
        super().__init__(brand, model, year)
        self.battery_capacity = battery_capacity

    def display_details(self):
        print(f"Electric Car: {self.year} {self.brand} {self.model}, Battery: {self.battery_capacity}")


ev = ElectricCar("Tata", "Nexon EV", 2023, "40.5 kWh")
ev.display_details()
```

### 5. Explain "Encapsulation" by creating private attributes using `__`.

```python
# Encapsulation means keeping data and methods together inside a class.
# Private attributes are made using double underscore __ before the name.
# This helps stop direct access from outside the class.

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def show_balance(self):
        print(f"{self.owner}'s balance is Rs.{self.__balance}")


account = BankAccount("Harsh", 5000)
account.show_balance()
# print(account.__balance)  # This will give an error because it is private.
```

### 6. Use `@property` to create getter and setter methods.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print("Salary should be positive")


emp = Employee("Rahul", 25000)
print(emp.salary)
emp.salary = 30000
print(emp.salary)
```

### 7. What is "Multiple Inheritance"? Provide a small code structure.

```python
# Multiple inheritance means one child class inherits from more than one parent class.

class Father:
    def father_quality(self):
        print("Good at planning")


class Mother:
    def mother_quality(self):
        print("Good at managing things")


class Child(Father, Mother):
    pass


child = Child()
child.father_quality()
child.mother_quality()
```

### 8. Create a class method using the `@classmethod` decorator.

```python
class College:
    college_name = "ABC College"

    @classmethod
    def change_college_name(cls, new_name):
        cls.college_name = new_name


print(College.college_name)
College.change_college_name("XYZ College")
print(College.college_name)
```

### 9. Explain the difference between a class method and a static method.

```python
# Class method:
# - Uses cls as the first parameter.
# - Can access and change class variables.
#
# Static method:
# - Does not use self or cls.
# - It behaves like a normal function, but it is kept inside a class.

class MathHelper:
    subject = "Maths"

    @classmethod
    def show_subject(cls):
        print("Subject:", cls.subject)

    @staticmethod
    def add(a, b):
        return a + b


MathHelper.show_subject()
print(MathHelper.add(10, 5))
```

### 10. Write a try-except-finally block to handle a `ZeroDivisionError`.

```python
try:
    number = 10
    result = number / 0
    print(result)
except ZeroDivisionError:
    print("Cannot divide a number by zero")
finally:
    print("This block always runs")
```
