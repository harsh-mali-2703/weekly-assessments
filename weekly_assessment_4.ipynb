 ###1. Explain the `break`, `continue`, and `pass` statements with examples

- `break` stops the loop completely.
- `continue` skips the current round and moves to the next one.
- `pass` does nothing. It is mostly used as a placeholder when we do not want to write code yet.

```python
# break example
for i in range(1, 6):
    if i == 4:
        break
    print(i)

print("---")

# continue example
for i in range(1, 6):
    if i == 3:
        continue
    print(i)

print("---")

# pass example
for i in range(1, 4):
    if i == 2:
        pass
    print(i)
```

###2. Convert a list of temperatures in Celsius to Fahrenheit using `map()`

```python
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9 / 5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

### 3. Find the second smallest element in a list

```python
numbers = [8, 3, 1, 6, 3, 9, 1, 5]

unique_numbers = sorted(set(numbers))

if len(unique_numbers) >= 2:
    print("Second smallest number is:", unique_numbers[1])
else:
    print("Second smallest number does not exist")
```

### 4. Write a program to check if a year is a leap year

```python
year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

### 5. How do you reverse a dictionary? Keys become values and values become keys

```python
student_marks = {
    "Amit": 85,
    "Riya": 92,
    "Neha": 78
}

reversed_dict = {value: key for key, value in student_marks.items()}

print(reversed_dict)
```

### 6. Demonstrate the use of `any()` and `all()` on a list of booleans

```python
answers = [True, True, False, True]

print("Any value is True:", any(answers))
print("All values are True:", all(answers))
```

### 7. Write a script to flatten a nested list

```python
nested_list = [[1, 2], [3, 4], [5, 6]]

flat_list = []

for small_list in nested_list:
    for item in small_list:
        flat_list.append(item)

print(flat_list)
```

### 8. Explain the difference between `append()` and `extend()` in lists

`append()` adds one item at the end of a list. If we append another list, it goes inside as one item.

`extend()` adds each item from another list separately.

```python
list1 = [1, 2, 3]
list1.append([4, 5])
print("After append:", list1)

list2 = [1, 2, 3]
list2.extend([4, 5])
print("After extend:", list2)
```

### 9. Write a script to find the longest word in a sentence

```python
sentence = "Python programming is interesting and useful"

words = sentence.split()
longest_word = max(words, key=len)

print("Longest word is:", longest_word)
```

### 10. Use `collections.Counter` to find the 3 most common items in a list

```python
from collections import Counter

items = ["apple", "banana", "apple", "orange", "banana", "apple", "grapes", "banana"]

count_items = Counter(items)
common_items = count_items.most_common(3)

print("Three most common items:", common_items)
```
