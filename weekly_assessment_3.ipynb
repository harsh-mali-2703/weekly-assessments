### 1. Write a while loop that terminates when a user types "exit".

```python
text = ""

while text != "exit":
    text = input("Type something, or type exit to stop: ")
    print("You typed:", text)

print("Loop ended")
```

### 2. Create a dictionary of 5 students and their marks; print only those who passed (>35).

```python
students = {
    "Aman": 72,
    "Riya": 34,
    "Karan": 56,
    "Neha": 29,
    "Vikram": 41
}

for name, marks in students.items():
    if marks > 35:
        print(name, "passed with", marks, "marks")
```

### 3. Use a list comprehension to square all numbers in a list.

```python
numbers = [2, 4, 6, 8, 10]
squares = [num ** 2 for num in numbers]

print(squares)
```

### 4. How do you merge two dictionaries in Python 3.9+?

```python
dict1 = {"name": "Harsh", "age": 21}
dict2 = {"city": "Mumbai", "course": "Python"}

merged_dict = dict1 | dict2

print(merged_dict)
```

### 5. Write a script to find the frequency of each character in a string.

```python
word = "banana"
frequency = {}

for ch in word:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1

print(frequency)
```

### 6. Demonstrate the enumerate() function in a for loop.

```python
subjects = ["Maths", "Science", "English"]

for index, subject in enumerate(subjects, start=1):
    print(index, subject)
```

### 7. Write a program to find the intersection of two sets.

```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

common_items = set_a.intersection(set_b)

print(common_items)
```

### 8. Use zip() to combine two lists into a dictionary.

```python
names = ["Aman", "Riya", "Karan"]
marks = [72, 88, 65]

result = dict(zip(names, marks))

print(result)
```

### 9. Create a 3x3 matrix using nested lists and access the middle element.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

middle = matrix[1][1]

print("Middle element is", middle)
```

### 10. Write a script to sort a dictionary by its values.

```python
scores = {
    "Aman": 72,
    "Riya": 88,
    "Karan": 65,
    "Neha": 91
}

sorted_scores = dict(sorted(scores.items(), key=lambda item: item[1]))

print(sorted_scores)
```
