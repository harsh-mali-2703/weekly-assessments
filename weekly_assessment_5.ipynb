### 1. Write a function that accepts variable-length arguments (*args).

```python
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(10, 20, 30))
print(add_numbers(5, 15, 25, 35))
```

### 2. Create a function that accepts keyword arguments (kwargs) and prints them.

```python
def show_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

show_details(name="Harsh", course="Python", week=5)
```

### 3. Write a recursive function to calculate the factorial of a number.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

### 4. Explain the difference between Global and Local scope with a code example.

```python
# A global variable is made outside a function, so it can be used in many places.
# A local variable is made inside a function, so it works only inside that function.

message = "I am global"

def scope_example():
    name = "I am local"
    print(message)
    print(name)

scope_example()
print(message)

# print(name)  # This would give an error because name is local.
```

### 5. Write a lambda function to sort a list of tuples based on the second element.

```python
students = [("Aman", 75), ("Riya", 92), ("Karan", 68), ("Neha", 85)]

sorted_students = sorted(students, key=lambda item: item[1])
print(sorted_students)
```

### 6. Create a "Filter" function to extract prime numbers from a list.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

numbers = [2, 3, 4, 5, 10, 11, 15, 17, 20]
prime_numbers = list(filter(is_prime, numbers))

print(prime_numbers)
```

### 7. What is a "Docstring" and how do you access it at runtime?

```python
# A docstring is a small description written inside triple quotes.
# It is usually written just after defining a function, class, or module.
# We can access it by using __doc__.

def greet_user():
    """This function prints a simple greeting."""
    print("Hello")

print(greet_user.__doc__)
```

### 8. Write a function that returns multiple values.

```python
def calculate(a, b):
    addition = a + b
    subtraction = a - b
    multiplication = a * b
    return addition, subtraction, multiplication

add, sub, mul = calculate(12, 4)
print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
```

### 9. Demonstrate the use of nonlocal in a nested function.

```python
def outer_function():
    count = 0

    def inner_function():
        nonlocal count
        count += 1
        print("Count is", count)

    inner_function()
    inner_function()

outer_function()
```

### 10. Create a function decorator that prints "Starting..." before any function runs.

```python
def starting_decorator(func):
    def wrapper():
        print("Starting...")
        func()
    return wrapper

@starting_decorator
def say_hello():
    print("Hello, the function is running now.")

say_hello()
```
